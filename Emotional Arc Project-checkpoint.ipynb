```python
# Get text files of books from Project Gutenberg

## Get the book titles
'''
1.
This section of the code is designed to navigate and scrape data from the Project Gutenberg Crime Fiction Bookshelf page:
"https://www.gutenberg.org/ebooks/bookshelf/28".
The objective is to extract the numerical book IDs from the hyperlinks associated with each book listed on the page.
To go to this website: "https://www.gutenberg.org/ebooks/bookshelf/68". 

2.
After collecting all book IDs, navigate to the next page if available.
The next page link is identified by a specific href attribute, typically linking to the subsequent set of books.

3. 
In a separate function, utilize the book_ids array to access the book text files.
Though, the text content I was trying to retrieve via inspecting elements is getting truncated in the development tools. So,
instead I fetched the entire page content directly rather than relying on element-specific selectors that might be affected by rendering limitations.
'''

import asyncio
from playwright.async_api import async_playwright
import re
```

```python
async def scrape_book_ids():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=False)  # Launches a visible browser
        page = await browser.new_page()
        
        # URL of the bookshelf page
        base_url = "https://www.gutenberg.org"
        url = f"{base_url}/ebooks/bookshelf/28"
        await page.goto(url)
        
        # List to store book IDs
        book_ids = []
        
        while True:  # Loop through each page
            # Wait for the content to load properly
            await page.wait_for_selector("li.booklink")
            
            # Fetch all book entries
            books = await page.query_selector_all("li.booklink")
            for book in books:
                # Check if the book is an audiobook or has a title ending with ")" because those are usually in diff. languages
                audiobook_check = await book.query_selector(".icon.icon_audiobook")
                title_element = await book.query_selector(".title")
                title_text = await title_element.inner_text() if title_element else ""
                
                if audiobook_check is None and not title_text.endswith(")"):
                    # If not an audiobook and title does not end with ")", get the link
                    link_element = await book.query_selector("a.link")
                    href = await link_element.get_attribute("href")
                    book_id = int(href.split('/')[-1])  # Extract and convert the ID to integer
                    book_ids.append(book_id)
            
            # Try to go to the next page
            next_link_element = await page.query_selector("a[title='Go to the next page of results.']")
            if next_link_element:
                next_page_url = await next_link_element.get_attribute("href")
                await page.goto(f"{base_url}{next_page_url}")
            else:
                break  # No more pages

        await browser.close()
        return book_ids
    
async def fetch_book_texts(book_ids):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=False)  # Launches a browser in headless mode
        page = await browser.new_page()

        # Base URL for fetching books
        base_url = "https://www.gutenberg.org/cache/epub/"

        # List to store the contents of each book
        book_texts = []

        for book_id in book_ids:
            # Construct the URL to access the book text
            url = f"{base_url}{book_id}/pg{book_id}.txt"
            response = await page.goto(url)
            
            # Fetch the complete document content as text
            text = await response.text()
            book_texts.append(text)
        
        await browser.close()
        return book_texts

def extract_and_clean_texts(texts):
    cleaned_texts = []
    book_titles = []
    
    for text in texts:
        # Search for a title in the format "Title: The War of the Worlds\n" before the start marker
        title_pre_start_match = re.search(r"Title: (.+)\n", text)
        if title_pre_start_match:
            book_title = title_pre_start_match.group(1).strip()
            book_titles.append(book_title)
        else:
            book_titles.append("Title not found")

        # Define the start and end patterns for the Gutenberg text
        start_pattern = re.escape("*** START OF THE PROJECT GUTENBERG EBOOK ") + "(.+?)" + re.escape(" ***")
        end_pattern = re.escape("*** END OF THE PROJECT GUTENBERG EBOOK ") + "(.+?)" + re.escape(" ***")
        
        # Search for start and end markers
        start_match = re.search(start_pattern, text)
        end_match = re.search(end_pattern, text)
        
        if start_match and end_match:
            # Extract the title from the start pattern (if needed)
            title = start_match.group(1)
            
            # Extract text between start and end markers
            extracted_text = text[start_match.end():end_match.start()]
            
            # Remove all occurrences of "\r" and "\n"
            cleaned_text = extracted_text.replace("\r\n", " ").replace("\n\r", " ").replace("\r", " ").replace("\n", " ")
            
            cleaned_texts.append(cleaned_text)
        else:
            # If no matches found, append an empty string or a specific error message
            cleaned_texts.append("No valid Gutenberg header/footer found.")

    return book_titles, cleaned_texts    
    
# Function to run the asyncio event loop
async def main():
    book_ids = await scrape_book_ids()
    texts = await fetch_book_texts(book_ids)
    book_titles, cleaned_texts = extract_and_clean_texts(texts)
    return book_titles, cleaned_texts
#     print("Scraped book IDs:")
#     for book_id in book_ids:
#         print(book_id)
```

```python
async def search_book_and_get_info(page):
    # Click on the first book link
    await page.click('.bookTitle[itemprop="url"]')

    # Get the content of the page
    content = await page.content()
    soup = BeautifulSoup(content, 'html.parser')
    
    # Extract the average rating
    average_rating_div = soup.find('div', class_="RatingStatistics__rating")
    average_rating = float(average_rating_div.get_text(strip=True)) if average_rating_div else 0.00

    # Extract the number of ratings
    ratings_span = soup.find('span', {"data-testid": "ratingsCount"})
    ratings_text = ratings_span.get_text(strip=True) if ratings_span else "0"
    number_of_ratings = int(re.sub("[^0-9]", "", ratings_text))

    # Return the results as a list
    return [average_rating, number_of_ratings]
```

```python
from transformers import AutoModelForSequenceClassification
from transformers import TFAutoModelForSequenceClassification
from transformers import AutoTokenizer, AutoConfig
import numpy as np
import pandas as pd
from scipy.special import softmax
from tqdm.notebook import tqdm
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
config = AutoConfig.from_pretrained(MODEL)
```

```python
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
def sentiment_score(text):
    output = model(**text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'negative' : scores[0],
        'neutral' : scores[1],
        'positive' : scores[2]
    }
    return (scores_dict)
```

```python
# Function to find the last period before the token limit
def find_last_space_before_limit(text, limit):
    # Tokenize the text and truncate to the limit
    tokens = tokenizer.tokenize(text)
    
    # Iterate over the tokens in reverse to find the last period within the limit
    period_indices = [i for i, token in enumerate(tokens) if token == ' ']
    # Find the last period that occurs before the token limit
    # If no such period is found, we split at the limit (might split a sentence)
    split_index = next((index for index in reversed(period_indices) if index < limit), limit)
    # Ensure the split index is within bounds
    split_index = min(split_index + 1, limit)  # +1 to include the period itself
    # Convert tokens back to text
    text_up_to_period = tokenizer.convert_tokens_to_string(tokens[:split_index])
    remaining_text = tokenizer.convert_tokens_to_string(tokens[split_index:]).strip()
    return text_up_to_period, remaining_text
```

```python
def score_all_texts(books):
    sentiment_book_scores = []

    for idx, transcript in enumerate(books):
#         print(transcript[:50])
        tokened_transcript = []

        # Process the transcript and tokenize based on the 512 token limit
        while transcript:
            chunk, transcript = find_last_space_before_limit(transcript, 512)
            tokened_chunk = tokenizer.encode_plus(
                chunk,
                max_length=512,
                padding='max_length',
                truncation=True,
                return_tensors='pt'
            )
            tokened_transcript.append(tokened_chunk)

        score = []

        for encoded_text in tokened_transcript:
            score.append(sentiment_score(encoded_text))
#             print(score[-1])
        print("Done with " + str(idx+1) + "/" + str(len(books)))

        sentiment_book_scores.append(score)
        
    return sentiment_book_scores
```

```python
#  count the number of peaks and troughs, the average distance between each peak, and the 
#  average peak-trough difference squared for each book

results = []

for idx, book in enumerate(net_sentiment_score):
    df = pd.DataFrame(book)
    net_sentiment = net_sentiment_score[idx]
    
    # Find peaks and troughs
    peaks, _ = find_peaks(net_sentiment)
    troughs, _ = find_peaks(-net_sentiment)
    
    # Calculate differences between each peak and each trough
    differences = [net_sentiment[p] - net_sentiment[t] for p in peaks for t in troughs]
    
    # Number of peaks and troughs
    num_peaks = len(peaks)
    num_troughs = len(troughs)
    
    # Average distance between peaks
    avg_peak_dist = np.mean(np.diff(peaks)) if num_peaks > 1 else 0
    
    # Average peak-trough difference squared
    avg_peak_trough_diff_squared = np.mean([d**2 for d in differences]) if differences else 0
    
    # Minimum and maximum peak-trough differences
    min_peak_trough_diff = min(differences) if differences else 0
    max_peak_trough_diff = max(differences) if differences else 0
    
    results.append({
        'Book': book_titles[idx],
        'Chapters': len(book),
        'Peaks': num_peaks,
        'Troughs': num_troughs,
        'Avg Peak Dist': avg_peak_dist,
        'Avg (Peak-Trough) squared': avg_peak_trough_diff_squared,
        'Min Peak Trough Diff': abs(min_peak_trough_diff),
        'Max Peak Trough Diff': max_peak_trough_diff        
    })

results_df = pd.DataFrame(results)
full_df = pd.merge (results_df, popularity_df[['Book', 'average_rating']], on='Book')
full_df
```

#### Model 1 (KNN)

This section of our analysis applies the K-Nearest Neighbors (KNN) algorithm to predict book ratings based on the sentiment distribution across chapters. Here is a detailed breakdown of the steps and methodologies employed:

##### Step 1: Normalization of Chapter Sentiments
Each book in our dataset contains sentiment scores for each chapter. To compare books of different lengths, we normalize the sentiment scores by the number of chapters:
- __Normalization ensures__ that each book's sentiment scores are scaled to a standard range, allowing for a fair comparison across books with varying numbers of chapters.

##### Step 2: Chi-Squared Test for Distribution Comparison
To assess the similarity in sentiment distribution across books, we employ the Chi-Squared test:
- __Chi-Squared Test:__ This statistical test compares the distribution of sentiment scores across normalized chapter indices among different books. It helps determine how closely related the sentiment distributions are across various books.

##### Step 3: K-Nearest Neighbors Algorithm
Using the results from the Chi-Squared test, we apply the K-Nearest Neighbors algorithm to predict the average ratings for books:
- __KNN Algorithm:__ This machine learning algorithm predicts the average rating of a new or unanalyzed book by considering the 'nearest' books—those with similar sentiment trends across chapters.

##### Step 4: Feature Engineering for KNN
To effectively implement the KNN, we transform our sentiment data into a suitable format:
- __Lambda Function:__ For each book, create a lambda function that encapsulates its sentiment distribution as a function of normalized chapter indices.
- __Vectorization:__ Convert these lambda functions into a vectorized form. This transformation is crucial for efficiently computing distances between books in the feature space.
- __Comparison:__ Instead of using histograms, our approach involves a piecewise function for each book. This method allows us to compare the exact points of the function, enhancing the accuracy of our predictions.

```python
# Normalize the length of sentiment score arrays
normalized_scores = [np.interp(np.linspace(0, 1, 100), np.linspace(0, 1, len(scores)), scores) for scores in net_sentiment_score]
```

```python
def create_functional_representation(scores):
    # Create a linear interpolation function for the scores
    return interp1d(np.linspace(0, 1, len(scores)), scores, kind='linear', fill_value="extrapolate")

functions = [create_functional_representation(scores) for scores in normalized_scores]

def compute_chi_squared_features(functions):
    num_books = len(functions)
    feature_matrix = np.zeros((num_books, num_books))
    x = np.linspace(0, 1, 100)
   
    for i in range(num_books):
        for j in range(num_books):
            observed = functions[i](x)
            expected = functions[j](x)
            chi2, _ = chisquare(observed, f_exp=expected)
            feature_matrix[i, j] = chi2
   
    # We can use sum of chi-squared values as features or another transformation
    return np.sum(feature_matrix, axis=1)

features = compute_chi_squared_features(functions)
```

```python
# Normalize features
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features.reshape(-1, 1))

# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(features_scaled, popularity_df['average_rating'], test_size=0.2, random_state=42)

# Train KNN
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)

# Predict ratings
predictions = knn.predict(X_test)
print("Predicted Ratings:", predictions)
```

```python
# Predict ratings for the test set
y_pred = knn.predict(X_test)

# Calculate error metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared Score:", r2)
```
